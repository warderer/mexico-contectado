```python
import pandas as pd
```

```python
# Cargando las rutas de los archivos a utilizar
mexico_conectado_path = 'data/11601Mexico_Conectado.xlsx'
indice_marginacion_path = 'data/IMM_2020.xls'
```

```python
# Cargando los archivos en DataFrames
try:
    df_mexico_conectado = pd.read_excel(mexico_conectado_path)
    print("Archivo de México Conectado cargado correctamente.")
except FileNotFoundError:
    print("Error: El archivo de México Conectado no se encontró en la ruta especificada.")

try:
    df_indice_marginacion = pd.read_excel(indice_marginacion_path, skiprows=5)
    # Eliminamos totales nacionales y fila en blanco
    df_indice_marginacion.drop([0,1], inplace=True)
    print("Archivo de índice de marginación cargado correctamente.")
except FileNotFoundError:
    print("Error: El archivo de índice de marginación no se encontró en la ruta especificada.")

```

```python
# Convertir todas las columnas a minúsculas
df_mexico_conectado.columns = df_mexico_conectado.columns.str.lower()
df_indice_marginacion.columns = df_indice_marginacion.columns.str.lower()
```

```python
# Ver los tipos de datos de las columnas
print(df_mexico_conectado.dtypes)
print(df_indice_marginacion.dtypes)
```

```python
# Renombrar Columnas de Estado y Municipio
df_indice_marginacion = df_indice_marginacion.rename(columns={'nom_ent':'estado','nom_mun':'municipio'})
```

```python
df_mexico_conectado.head()
```

```python
# Número de registros del DataFrame de Indice de Marginación
len(df_indice_marginacion)
```

```python
# Número de registros del DataFrame de México Conectado
len(df_mexico_conectado)
```

```python
# Eliminación de columnas con valores vacios
df_indice_marginacion = df_indice_marginacion.dropna(how='any')
df_mexico_conectado = df_mexico_conectado.dropna(how='any')
```

```python
# Número de registros del Indice de Marginación después de ejecutar la limpieza
len(df_indice_marginacion)
```

```python
# Número de registros de México Conectado después de ejecutar la limpieza
```

```python
len(df_mexico_conectado)
```

```python
# Comprobación del número de municipios únicos en Mexico Conectado
df_mexico_conectado['municipio'].nunique()
```

```python
# Cambiar las columnas de número decimal a números enteros
df_indice_marginacion['cve_mun'] = df_indice_marginacion['cve_mun'].astype(int)
df_indice_marginacion['pob_tot'] = df_indice_marginacion['pob_tot'].astype(int)
```

```python
 df_indice_marginacion.head(10)
```

```python
# En México Conectado: Agrupar por 'municipio', calcular el promedio de ancho de banda
# y contar el número de registros por municipio
df_promedio_por_municipio = df_mexico_conectado.groupby('municipio').agg({
    'ancho_banda_bajada': 'mean',
    'ancho_banda_subida': 'mean',
    'municipio': 'count'
}).rename(columns={'municipio': 'conteo_registros'}).reset_index()

# Visualizar el resultado
print(df_promedio_por_municipio)
```

```python
# Convertir la columna 'municipio' de ambos DataFrames a mayúsculas
df_indice_marginacion['municipio'] = df_indice_marginacion['municipio'].str.upper()
df_promedio_por_municipio['municipio'] = df_promedio_por_municipio['municipio'].str.upper()
```

```python
# Hacer el left join para unir ambos DataFrames en uno solo
df_merged = pd.merge(df_indice_marginacion, df_promedio_por_municipio, on='municipio', how='left')
print(df_merged)
```

```python
len(df_merged)
```

```python
df_merged.head(20)
```

```python
# Al mezclar las tablas con Left Join, quedarán campos inválidos en Mexico Conectado.
# Reemplazar NaN por 0 en df_merged
df_merged = df_merged.fillna(0)
df_merged.head(20)
```

```python
# Ver los tipos de datos de las columnas
print(df_merged.dtypes)
```

```python
# Convertir la columna 'conteo_registros' de float64 a int64
df_merged['conteo_registros'] = df_merged['conteo_registros'].astype('int64')
df_merged.head()
```

```python
# Renombrar columnas de la tabla resultante a más legibles
df_merged = df_merged.rename(columns=
    {'cve_ent':'clave_entidad','cve_mun':'clave_municipio','pob_tot':'poblacion_total','im_2020':'indice_marginacion_2020',
     'gm_2020':'grado_marginacion_2020','imn_2020':'indice_marginacion_normalizado_2020','unnamed: 17':'lugar_nacional'})
```
